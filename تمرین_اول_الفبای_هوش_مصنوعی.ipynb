```python
from matplotlib import pyplot as plt
```

```python
import cv2
```

```python
import numpy as np
```

```python
image= cv2.imread("saman.jpg")
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

```python
model=cv2.CascadeClassifier("model.xml")
```

```python
faces =model.detectMultiScale(image)
```

```python
print(faces)
```

```python
image.shape
```

```python
x = faces[0][0]
y = faces[0][1]
a = faces[0][2]
b = faces[0][3]
```

```python
image=cv2.rectangle(image,(x,y),(x+a,y+b),(0,255,0),5)
```

```python
# نمایش تصویر
plt.imshow(image)
```
